```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
# implementing decision_tree_classifer 

x = df.drop('Class',axis=1)
y = df['Class']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.3)

ms = MinMaxScaler()
scaled_xtrain = ms.fit_transform(x_train,y_train)
scaled_xtest = ms.transform(x_test)

dt_class = DecisionTreeClassifier()

param_grid = {
    'criterion':['gini','entropy'],
    'max_depth':[5,6,7]
}


grid_scv = GridSearchCV(estimator=dt_class,param_grid=param_grid,cv=7)

grid_scv.fit(scaled_xtrain,y_train)
```

```python
best_model=grid_scv.best_estimator_
print(f'Best_params : {grid_scv.best_params_}')

y_predicted = best_model.predict(scaled_xtest)
y_predicted
```

```python
acc_score = accuracy_score(y_test,y_predicted)
acc_score
```

```python
from sklearn import metrics

print('MAE:',metrics.mean_absolute_error(y_test,y_predicted))
print('MSE:',metrics.mean_squared_error(y_test,y_predicted))
print('RMSE:',np.sqrt(metrics.mean_absolute_error(y_test,y_predicted)))
```
